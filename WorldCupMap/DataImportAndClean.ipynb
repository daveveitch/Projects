# Import Census Data

```python
census_data_original = pd.read_csv('98-401-X2016043_English_CSV_data.csv')
```

# Import Toronto Census Tracts

```python
# Import census tracts from csv and convert it to file format compatible with census data
toronto_census_tracts = pd.read_csv('TorontoCensusTracts.csv',dtype=object,header=None)
toronto_census_tracts = ('535'+toronto_census_tracts[[0]])
toronto_census_tracts[0] = toronto_census_tracts[0].str.replace('.','') 
toronto_census_tracts = toronto_census_tracts.rename(columns={0:'ALT_GEO_CODE'})
toronto_census_tracts
```

# Import Features I am interested in

```python
world_cup_nationalities = pd.read_csv('TeamEthnicOrigin.csv')
```

```python
world_cup_nationalities
```

```python
# Pick out section from data
census_data_filter = census_data_original[(census_data_original['ALT_GEO_CODE']>=int(min(toronto_census_tracts['ALT_GEO_CODE'])))&
                                         (census_data_original['ALT_GEO_CODE']<=int(max(toronto_census_tracts['ALT_GEO_CODE'])))]

# Merge dataframes together to get only required census tracts and relevant fields
census_data_filter = pd.merge(census_data_filter,toronto_census_tracts,on='ALT_GEO_CODE',how='left')
census_data_filter = pd.merge(census_data_filter,world_cup_nationalities,on='Member ID: Profile of Census Tracts (2247)',how='inner')
census_data_filter
```

# Write Cleaned DF to a CSV File

```python
# Create first row of a dataframe that will contain each census tract and
# % populations of each ethnic group

# Creates a single dataframe with one row, census tract as row name and
# each column % of individuals in census tract of certain nationalities
eth_by_tract = census_data_filter[census_data_filter['ALT_GEO_CODE']==535000100]
eth_by_tract = eth_by_tract[['ALT_GEO_CODE','DIM: Profile of Census Tracts (2247)','Dim: Sex (3): Member ID: [1]: Total - Sex']]

first_entry_index = list(eth_by_tract.index)[0]
tract = eth_by_tract['ALT_GEO_CODE'][first_entry_index]

eth_by_tract = eth_by_tract.rename(columns={'Dim: Sex (3): Member ID: [1]: Total - Sex':tract})
eth_by_tract = eth_by_tract.set_index('DIM: Profile of Census Tracts (2247)')
eth_by_tract = eth_by_tract[[tract]]
eth_by_tract[tract] = eth_by_tract[tract].astype('int64')/int(eth_by_tract[tract]['Total - Ethnic origin for the population in private households - 25% sample data'])
eth_by_tract = eth_by_tract.T

# iterates through other census tracts and builds a dataframe with every census
# tract and ethnicities
for i in toronto_census_tracts['ALT_GEO_CODE'][1:]:
    single_tract = census_data_filter[census_data_filter['ALT_GEO_CODE']==int(i)]
    single_tract = single_tract[['ALT_GEO_CODE','DIM: Profile of Census Tracts (2247)','Dim: Sex (3): Member ID: [1]: Total - Sex']]
    first_entry_index = list(single_tract.index)[0]
    tract = single_tract['ALT_GEO_CODE'][first_entry_index]
    
    single_tract = single_tract.rename(columns={'Dim: Sex (3): Member ID: [1]: Total - Sex':tract})
    single_tract = single_tract.set_index('DIM: Profile of Census Tracts (2247)')
    single_tract = single_tract[[tract]]
    single_tract.loc[(single_tract[int(i)]=='x')]=0    
    single_tract[tract] = single_tract[tract].astype('int64')/int(single_tract[tract]['Total - Ethnic origin for the population in private households - 25% sample data'])
    single_tract = single_tract.T
  
    eth_by_tract = eth_by_tract.append(single_tract)

# deletes census tracts with no data    
eth_by_tract = eth_by_tract.fillna(0)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
eth_by_tract = eth_by_tract[eth_by_tract['Total - Ethnic origin for the population in private households - 25% sample data']>0]

# writes dataframe to a csv file
eth_by_tract.to_csv('EthByTract.csv',index_label='ALT_GEO_CODE')
```
